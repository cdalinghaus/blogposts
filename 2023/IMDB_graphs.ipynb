The internet is home to a plethora of databases catering to various fields of interest. One such valuable treasure trove of data is the IMDb, the "Internet Movie Database." In this article, we're going to navigate through this vast sea of data, specifically focusing on the intricate web of connections between actors. The objective? To discover the hidden clusters and networks among the most successful actors based on 2021 box office revenues.

#### Understanding the IMDb
IMDb, an Amazon subsidiary since 1998, serves as an online repository for movie and media metadata. Registered users can rate and write reviews for movies and series on a scale of 1 to 10, providing a rich source of data for sentiment analysis. IMDb also shares a portion of its data publicly, which is precisely what we utilized for our exploration.

#### Unraveling the Network of Actors
We often associate certain movie genres or franchises with a specific group of actors. For instance, Action films or Comic adaptations often feature a select group of performers who seem to collaborate frequently. Our hypothesis is that there exist groups among the top actors who often work together, but seldom with actors from other groups.

To validate our hypothesis, we created an unweighted graph featuring the top 100 actors of 2021 based on box office revenues. Each actor (node) was linked (edge) to every other actor they had worked with in three or more projects. We used "[Gephi](https://gephi.org/)," a graph visualization software, along with the ForceAtlas2 algorithm, to create a meaningful spatial arrangement of the nodes.

This network analysis unearthed several interesting clusters. Here's what we found:

#### The Marvel Cluster


The central cluster is characterized by a dense network of actors who all play roles within the Marvel Universe. These films are known for their interconnected plots and crossover appearances, explaining the strong intra-cluster connections.

#### The Fast & Furious / Guardians of the Galaxy / WWE Cluster



This cluster is interesting as it showcases a connection anchored by Dwayne Johnson. This cluster also has ties to the Marvel Cluster and features many collaborations between Dwayne Johnson, John Cena, and Dave Bautista, who all have roots in WWE (World Wrestling Entertainment). Furthermore, several actors in this cluster belong to the "Fast & Furious" and "Guardians of the Galaxy" franchises.

#### The Star Wars Cluster



The Star Wars Cluster is less interconnected than the previous ones. Still, many direct connections exist between the actors, all of whom have roles in the "Star Wars" franchise.

#### The Cluster of the Older Generation



Here, Matt Damon stands out as a well-connected node, indicative of his collaborations with other actors in this cluster. This group comprises actors who have enjoyed successful careers over several decades, distinguishing them from younger generations.

#### The All-Rounder: Samuel L. Jackson



Finally, the most well-connected actor in the network is none other than Samuel L. Jackson. He shares multiple collaborations with actors from all clusters, owing to his roles in both the Marvel and Star Wars franchises and his decades-long successful career.

#### Final Thoughts

Interestingly, all strongly interconnected actors are men, despite 34 women being listed in the top 100 actors for 2021. Therefore, our initial hypothesis of group formation among top actors holds true, but primarily for male actors. Moreover, group formation doesn't necessarily imply that all actors in a group collaborate with each other; sometimes, individual actors serve as a connective hub. Finally, actors can simultaneously belong to several groups, underlining the versatile and dynamic nature of their professional networks.
