```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
# Initialize a directed graph
G = nx.DiGraph()
```

```python
for line in lines.split("\n"):
    if line.startswith("create_node"):
        # Extract the node name and add to the graph
        node_name = line[13:-3]
        G.add_node(node_name)

    if line.startswith("create_edge"):
        # Extract the edge details and add to the graph
        edge_details = line[12:-2].split(', ')
        relationship = edge_details[0][1:-1]
        from_node = edge_details[1][1:-1]
        to_node = edge_details[2][1:-1]
        G.add_edge(from_node, to_node, label=relationship)
```

```python
# Optional: Visualize the graph
pos = nx.kamada_kawai_layout(G)
nx.draw(G, pos, with_labels=True, font_size=8)
edge_labels = nx.get_edge_attributes(G, 'label')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_size=8)
plt.show()
```
