Unlike many other shows from the same era, like Scrubs, Friends was originally filmed using Panavision Panaflex cameras. Why is this a big deal? Well, these cameras shot in the 16:9 format but the episodes were cropped to 4:3 in post-production, which was the standard TV format back then.

#### So, What's The Big Deal?
By keeping the original film rolls, the creators were able to re-release Friends in its original 16:9 widescreen glory. But that's where things get interesting. The show now has what you might call "odd framing," featuring unused space on the right and left sides of the screen.

#### Our Experiment
To bring this unique framing into focus, we used a Convolutional Neural Network (CNN) implemented in PyTorch to detect faces and their positions in each frame. We repeated this for each frame in an episode and mapped the results onto a heatmap.

After applying our algorithm to three random episodes from each season of Friends, we found a pretty consistent pattern. Except for one outlier, the framing is largely uniform across episodes.

#### The Exception: Season 10, Episode 8
This particular episode bucks the trend. Our heatmap expands almost to the bottom of the frame. Why? Because in one scene, several characters peek their heads through a slightly opened door, stacked on top of each other. But let's be clearâ€”this episode is an anomaly. In general, the show sticks to its usual, conventional framing.

#### How Does Friends Compare to Native 16:9 Films?
To see if Friends was wasting the extra screen space, we compared its heatmaps to those of two native 16:9 films: "Downton Abbey (2019)" and "Lost in Translation (2003)."

The hotspots in these films are noticeably broader than those in Friends, and there's more separation between faces. This is likely due to the 180-degree rule in filmmaking, which guides the camera angles relative to the actors. What's more, these films make use of the screen space beyond the 4:3 boundary, something Friends does not do.

#### What Does This Mean?
If you're watching the 16:9 re-release of Friends on Netflix, you're not getting the full visual experience you might expect. The extra screen space outside the 4:3 frame is pretty much wasted, confirming our initial suspicion. So, while the widescreen version offers some new angles, it's not offering a substantially different visual experience.
