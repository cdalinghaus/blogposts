Scrolling through algorithmically curated feeds has become a go-to activity for millions of people online. But let’s be real: most of us are starting to realize just how much of a time sink this is. Blocking short-form content on your phone, like TikTok or Instagram reels, is a great first step. But I’d argue the problem goes deeper. Take reddit for example. While reddit usually is a good way to get useful information (you are guaranteed to a reasonable degree to be reading "an actual human thought" [Quote from [https://news.ycombinator.com/item?id=21403294](https://news.ycombinator.com/item?id=21403294)]), the homepage is a hellhole of attention-bating low-value content. Multiple times now have I been reading up on some persons trouble with their landlord on /r/renting and on some other persons trouble with their tenants. Which would be okay if it were a casual conversation with a friend, but coming from strangers on the internet? Come on!

I was surprised that no comprehensive firefox plugin exists that just blocks all of the algorithmically curated feeds among the sites I frequent. So from now on I will be curating a blocklist for Ublock Origin that takes care of removing these feeds without outright blocking the sites. Pull requests welcome:

https://github.com/cdalinghaus/ublock-feeds-blocklist

Also, if you're interested in a deep dive into the history of these algorithmically curated feeds I am recommending to read this: [https://medium.com/@annawchung/news-feeds-old-content-a-brief-history-of-algorithmically-curated-feeds-on-facebook-and-twitter-85b5e5d8e30a](https://medium.com/@annawchung/news-feeds-old-content-a-brief-history-of-algorithmically-curated-feeds-on-facebook-and-twitter-85b5e5d8e30a), as well as reading through the old [blog.facebook.com](https://web.archive.org/web/20061205110504/http://blog.facebook.com/) blog on the wayback machine. It's a fascinating piece of recent internet history.
